# Train/Test Split

This notebook splits the cleaned dataset into train, validation, and test sets based on temporal cutoffs, then saves each partition.

```python
import pandas as pd
import os
```

```python
df = pd.read_parquet("../data/bronze/clean_dataset.parquet", engine='pyarrow')
df.shape
```

The dataset goes from the 1st of January 2024 to the 30th December 2024, I will split it following 75% train, 15% val, 10% test.

```python
df = df.sort_values("date")

train_end = "2024-09-30"
val_end = "2024-11-30"

train_df = df[df["date"] <= train_end]
val_df   = df[(df["date"] > train_end) & (df["date"] <= val_end)]
test_df  = df[df["date"] > val_end]
```

```python
split_info = pd.DataFrame({
    "split": ["Train", "Validation", "Test"],
    "rows": [len(train_df), len(val_df), len(test_df)],
    "percentage": [round(len(train_df)/len(df), 2), round(len(val_df)/len(df), 2),round(len(test_df)/len(df), 2)]
})
split_info
```

## Save dataset splits

```python
output_dir = "../data/bronze"
os.makedirs(output_dir, exist_ok=True)

train_df.to_parquet(os.path.join(output_dir, "train.parquet"), engine='pyarrow', index=False)
val_df.to_parquet(os.path.join(output_dir, "val.parquet"), engine='pyarrow', index=False)
test_df.to_parquet(os.path.join(output_dir, "test.parquet"), engine='pyarrow', index=False)

print(f"Saved to {output_dir}/")
```
