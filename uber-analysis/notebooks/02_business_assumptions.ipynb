## Time coverage

Is the dataset really covering the whole year 2024?

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
full_range = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
daily_counts = df.groupby('Date').size()
missing_days = full_range.difference(daily_counts.index)
print("Days without data:", missing_days)
print("Number of missing days:", len(missing_days))
```

The dataset includes all days of the year except December 31st. It is important to note this omission and, if possible, obtain the data for December 31st to ensure the model covers the full year
